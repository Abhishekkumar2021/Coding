I have been given a rotated sorted array and i want to visualise the algorithm for finding the pivot element.
Example : arr = [4,5,6,7,8,9,10,0,1,2] pivot = 6 (0 based indexing)

Algorithm :
def findPivot(arr):
    low = 0
    high = len(arr)-1
    while low <= high:
        mid = (low+high)//2
        if mid + 1 <= high and arr[mid] > arr[mid+1]:
            return mid
        elif mid - 1 >= low and arr[mid] < arr[mid-1]:
            return mid-1
        elif arr[mid] > arr[high]:
            low = mid + 1
        else:
            high = mid - 1
    return -1

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# set grid spacing to 1
plt.xticks(range(0, 11, 1))

arr = [4,5,6,7,8,9,10,0,1,2]

# Algorithm to find the pivot
left = 0
right = len(arr) - 1
while left <= right:
    mid = (left + right) // 2
    # Plot the points of interest & label them
    plt.plot(arr, marker='o', color='b')
    plt.plot(left, arr[left], marker='o', color='r')
    plt.plot(right, arr[right], marker='o', color='y')
    plt.plot(mid, arr[mid], marker='o', color='g')  
    plt.text(left - 0.5, arr[left] + 0.5, 'left')
    plt.text(right - 0.5, arr[right] + 0.5 , 'right')
    plt.text(mid - 0.5, arr[mid] + 0.5, 'mid')
    plt.show()
    if mid + 1 <= right and arr[mid] > arr[mid + 1]:
        print(mid)
        break
    elif mid - 1 >= left and arr[mid] < arr[mid - 1]:
        print(mid - 1)
        break
    elif arr[mid] <= arr[left]:
        right = mid - 1
    else:
        left = mid + 1
    
    
    
    
```
